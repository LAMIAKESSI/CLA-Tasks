```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import norm
```

```python
# read the image file 

imgMat = mpimg.imread('cat.jpg')
plt.imshow(imgMat)
```

```python
imgMat.shape
```

```python
# transpose the matrix

imgTrans=imgMat.T
imgTrans.shape
```

```python
# check if the matrix is diagonal

def isDiagonal (matrix):
    isDiag = True
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if (matrix[i][j] != 0) and (i != j):
                isDiag = False
    return isDiag
```

```python
mat = imgTrans[0,:,:]
mat.shape
isDiagonal(mat)
```

```python
# check if the matrix is triangular first method

def isTriangonal (matrix):
    isUpTriang = True
    isLowTriang = True
    for i in range(1,len(matrix)):
        for j in range(i):
            if (matrix[i][j] != 0):
                isUpTriang = False
                break
    for i in range(len(matrix)):
        for j in range(1+i,len(matrix[0])):
            if (matrix[i][j] != 0):
                isLowTriang = False
                break
    
    return (isUpTriang, isLowTriang)
    
```

```python
mat = imgTrans[0,:,:]
isTriangonal(mat)
```

```python
# check if the matrix is triangonal second method

def isTriangular (matrix):
    LowTriang = np.allclose(matrix, np.tril(matrix)) # check if lower triangular
    UpTriang = np.allclose(matrix, np.triu(matrix)) # check if upper triangular
    return (UpTriang, LowTriang)
```

```python
A = np.array([[2,0,0],[0,2,0],[0,0,0]])
A
isTriangular(A)
```

```python
# check if the matrix is an identity matrix

def isIdentity (matrix):
    isIden = True
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if (matrix[i][j] != 0) and (i != j):
                isIden = False
                break
            elif (i == j) and (matrix[i][j] != 1):
                isIden = False   
                break     
              
    return isIden
                
```

```python
mat = imgTrans[0,:,:]
isIdentity(mat)
```

```python
# chaeck if a given matrix A is the inverse of matrix B

def isInverse (A,B):
    
    inverse = np.linalg.inv(A)
    if inverse.all() == B.all() :
        return True
    else :
        return False
    
```

```python
matA = imgTrans[0,:100,:100] # matA must be square
matB = np.linalg.inv(matA)
isInverse(matA, matB)
```

```python
matA = imgTrans[0,:100,:100] # matA must be square
matC = np.array([[1, 8, 9],[2, 0, 4],[1, 0, 6]])
isInverse(matA, matC)
```
