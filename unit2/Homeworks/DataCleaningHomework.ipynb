## 3) Detecting the possible outliers

```python
df.describe()
```

```python
numerics_types = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
new_df = df.select_dtypes(include=numerics)
new_df
```

```python
# Calculate the z-scores
z_scores = stats.zscore(new_df, nan_policy='omit')
z_scores
```

```python
# Convert to absolute values
abs_z_scores = np.abs(z_scores)

# Select data points with a z-scores above or below 3
filtered_entries = (abs_z_scores < 3).all(axis=1)

# Filter the dataset
df_outliers = new_df[filtered_entries]
df_outliers
```

## 6) Counting tags and finding the most common ones

```python
# spilting the text

text=""
for i in df.tags :
    text=text+str(i)
    
tag = text.split("|")
tag_df = pd.DataFrame(tag)
tag_df.columns=['tag']
tag_df
```

```python
# counting tags number

tag_nb = len(tag_df['tag'].unique())
print("There are", tag_nb,"tags.")
    
```

```python
# The most common tag
tag_df['tag'].mode()
```

```python
# A list of the most common tags
common_tags = tag_df.value_counts()
```
